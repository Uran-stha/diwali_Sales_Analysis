```python
#Total Number of Orders from top 10 States

# Prepare data
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Generate a palette with one color per state
colors = sns.color_palette("husl", len(sales_state))

# Plot
sns.set(rc={'figure.figsize': (18, 5)})
sns.barplot(data=sales_state, x='State', y='Orders', hue='State', palette=colors, legend=False)

plt.title('Total Number of Orders from Top 10 States')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Group by state and calculate total sales
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Define a distinct color for each bar
colors = sns.color_palette("husl", len(sales_state))

# Set plot size
sns.set(rc={'figure.figsize': (15, 5)})

# Create the bar plot
ax = sns.barplot(data=sales_state, x='State', y='Amount', hue='State', palette=colors, legend=False)

# Add data labels in millions (e.g., $1.2M)
for p in ax.patches:
    height = p.get_height()
    label = f'{height/1_000_000:.1f}M'  # Convert to millions and format
    ax.annotate(label,
                (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom', fontsize=10, fontweight='bold')

# Add chart formatting
plt.title('Total Sales Amount from Top 10 States')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*
```

```python
# Number of unique categories
num_categories = df['Product_Category'].nunique()

# Generate a color palette (you can try 'Set3', 'Paired', or 'husl')
colors = sns.color_palette("Set3", num_categories)

# Set figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Create the countplot with hue and custom palette
ax = sns.countplot(data=df, x='Product_Category', hue='Product_Category', palette=colors, legend=False)

# Add count labels on top
for bar in ax.containers:
    ax.bar_label(bar, fontsize=10, fontweight='bold')

# Add formatting
plt.title('Count by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# 1. Group and get top 10 product categories by sales amount
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum() \
                .sort_values(by='Amount', ascending=False).head(10)

# 2. Create a color palette with the same number of colors as bars
colors = sns.color_palette("Set3", len(sales_state))  # Try "husl", "Paired", or "Dark2" too

# 3. Set figure size
sns.set(rc={'figure.figsize': (20, 5)})

# 4. Plot with hue to apply unique colors, and disable legend
ax = sns.barplot(
    data=sales_state,
    x='Product_Category',
    y='Amount',
    hue='Product_Category',
    palette=colors,
    dodge=False,       # avoids offsetting bars by hue
    legend=False       # suppress redundant legend
)

# 5. Optional: Add labels on top of bars, formatted in millions
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height / 1_000_000:.1f}M',
                (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom', fontsize=10, fontweight='bold')

# 6. Format the plot
plt.title('Top 10 Product Categories by Sales Amount')
plt.xlabel('Product Category')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()

# 7. Show the plot
plt.show()
```

```python
# Step 1: Prepare data
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum() \
                .sort_values(by='Orders', ascending=False).head(10)

# Step 2: Create a color palette
colors = sns.color_palette("Set3", len(sales_state))  # Try "Paired", "husl", or "Dark2" as alternatives

# Step 3: Set figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Step 4: Create barplot with hue and custom colors
ax = sns.barplot(
    data=sales_state,
    x='Product_ID',
    y='Orders',
    hue='Product_ID',
    palette=colors,
    dodge=False,     # Keeps bars aligned
    legend=False     # Hides redundant legend
)

# Step 5: Add labels on top of bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}',
                (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom', fontsize=10, fontweight='bold')

# Step 6: Customize plot
plt.title('Top 10 Products by Number of Orders')
plt.xlabel('Product ID')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.tight_layout()

# Step 7: Show the plot
plt.show()
```

```python
*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*
```
